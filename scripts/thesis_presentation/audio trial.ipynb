```python
import neo
from scipy.io import wavfile
fs = 30000

path = r"C:\Users\Admin\Desktop\audio trial\testaudio_260120_103921\info.rhd"
reader = neo.io.IntanIO(path)
seg = reader.read_segment()


signal = seg.analogsignals[0].rescale('V').magnitude


sound = seg.analogsignals[2].magnitude.flatten() 

times = seg.analogsignals[0].times


# Save sound data as .wav file
wavfile.write('audio_output.wav', int(fs), sound.astype('float32'))
```

```python
import numpy as np

digital = seg.analogsignals[3].magnitude.flatten()
binary = digital == 1
fs = seg.analogsignals[3].sampling_rate.magnitude

mid = len(binary) // 2
offset = int(0.005 * fs)

# First half
indices_1 = np.where(binary[:mid])[0]
first_1, last_1 = indices_1[0], indices_1[-1] + offset

# Second half
indices_2 = np.where(binary[mid:])[0] + mid
first_2, last_2 = indices_2[0], indices_2[-1] + offset

print(f"Trial 1: first={first_1}, last={last_1}")
print(f"Trial 2: first={first_2}, last={last_2}")


audio = sound[first_1:last_1]  # For trial 1
wavfile.write('audio_trial_1.wav', int(fs), audio.astype('float32')) # excluded folder
audio = sound[first_2:last_2]  # For trial 2
wavfile.write('audio_trial_2.wav', int(fs), audio.astype('float32'))
```
